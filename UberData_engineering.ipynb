```python
# Importing the required library and reading the file in datafrmae
import pandas as pd
tripDetail = pd.read_csv("uber_data.csv")
tripDetail
```

```python
tripDetail.info()
```

```python
# Converting the date time object type to time
tripDetail['tpep_pickup_datetime'] = pd.to_datetime(tripDetail['tpep_pickup_datetime'])
tripDetail['tpep_dropoff_datetime'] = pd.to_datetime(tripDetail['tpep_dropoff_datetime'])
```

```python
tripDetail
```

```python
# Checking how many records we are dropping
(tripDetail[['tpep_pickup_datetime','tpep_dropoff_datetime']].drop_duplicates()).count()
```

```python
# Creating new dataframe after removing the duplicates
dateTime_dim = tripDetail[['tpep_pickup_datetime','tpep_dropoff_datetime']].drop_duplicates().reset_index(drop=True)
dateTime_dim.count()
```

```python
# Extracting hour day month year details from time details of pickup and dropoff
dateTime_dim['pick_hour'] = dateTime_dim['tpep_pickup_datetime'].dt.hour
dateTime_dim['pick_day'] = dateTime_dim['tpep_pickup_datetime'].dt.day
dateTime_dim['pick_month'] = dateTime_dim['tpep_pickup_datetime'].dt.month
dateTime_dim['pick_year'] = dateTime_dim['tpep_pickup_datetime'].dt.year
dateTime_dim['pick_weekOfDay'] = dateTime_dim['tpep_pickup_datetime'].dt.weekday

dateTime_dim['dropoff_hour'] = dateTime_dim['tpep_dropoff_datetime'].dt.hour
dateTime_dim['dropoff_day'] = dateTime_dim['tpep_dropoff_datetime'].dt.day
dateTime_dim['dropoff_month'] = dateTime_dim['tpep_dropoff_datetime'].dt.month
dateTime_dim['dropoff_year'] = dateTime_dim['tpep_dropoff_datetime'].dt.year
dateTime_dim['drop_weekOfDay'] = dateTime_dim['tpep_dropoff_datetime'].dt.weekday

dateTime_dim
```

```python
# Generating index for new column for primary key
dateTime_dim['datetime_id'] = dateTime_dim.index
```

```python
dateTime_dim
```

```python
# Currently dateTime_id is at the last of column. Regraming it to make it first
dateTime_dim.insert(0,'dateTime_id',dateTime_dim.index)
```

```python
# as old datetime_id is at last i am going to drop it as it is not required
dateTime_dim.drop('datetime_id',axis=1,inplace=True)
```

```python
dateTime_dim
```

```python
# Dropping duplicates for passenger count to create passenger dimension table
passenger_count_dim = tripDetail[['passenger_count']].drop_duplicates().reset_index(drop=True)
passenger_count_dim.insert(0,'passenger_count_id',passenger_count_dim.index)
passenger_count_dim
```

```python
# Dropping duplicates for trip 
trip_dim = tripDetail[['trip_distance']].drop_duplicates().reset_index(drop=True)
trip_dim.insert(0,'trip_id',trip_dim.index)
trip_dim
```

```python
#Mapping rate code in diension table of rate
rate_code_type = {
    1: "Standard rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated fare",
    6: "Group ride"
}

rate_code_dim = tripDetail[['RatecodeID']].drop_duplicates().reset_index(drop=True)
rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)
rate_code_dim.insert(0,'rate_code_id',rate_code_dim.index)
rate_code_dim
```

```python
#Mapping payment type to dimension table
payment_type_name = {
    1: "Credit card",
    2: "Cash",
    3: "No charge",
    4: "Dispute",
    5: "Unknown",
    6: "Voided trip"
}

payment_code_dim = tripDetail[["payment_type"]].drop_duplicates().reset_index(drop=True)
payment_code_dim["payment_type_name"] = payment_code_dim['payment_type'].map(payment_type_name)
payment_code_dim.insert(0,"payment_type_id",payment_code_dim.index)
payment_code_dim
```

```python
#Pickup details dimension table
pickup_dim = tripDetail[["pickup_longitude","pickup_latitude"]].drop_duplicates().reset_index(drop=True)
pickup_dim.insert(0,"pickup_id",pickup_dim.index)
pickup_dim
```

```python
#drop off details dimension table
dropoff_dim = tripDetail[["dropoff_longitude","dropoff_latitude"]].drop_duplicates().reset_index(drop=True)
dropoff_dim.insert(0,"dropoff_id",dropoff_dim.index)
dropoff_dim
```

```python
# Getting all the column names in dataframe
column_headers = list(tripDetail.columns.values)
print("The column header ;" , column_headers)
```

```python
# Creating fact table based upon dimension table
fact_table = tripDetail.merge(dropoff_dim,on=["dropoff_longitude","dropoff_latitude"]) \
                        .merge(pickup_dim,on=["pickup_longitude","pickup_latitude"]) \
                        .merge(payment_code_dim,on="payment_type") \
                        .merge(rate_code_dim,on='RatecodeID') \
                        .merge(trip_dim,on='trip_distance') \
                        .merge(dateTime_dim,on=['tpep_pickup_datetime','tpep_dropoff_datetime']) \
                        .merge(passenger_count_dim,on='passenger_count') \
                        [['VendorID', 'dateTime_id', 
                          'passenger_count_id', 'trip_id', 'pickup_id' , 
                          'rate_code_id', 'store_and_fwd_flag', 'dropoff_id', 
                          'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 
                          'improvement_surcharge', 'total_amount']]
```
